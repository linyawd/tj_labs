### 1.2 Функція розподілу

Виведіть значення функції розподілу від'ємного біноміального розподілу з параметрами $n, p$ в точках вектора $x$.

```python
# Місце для Вашого коду

from scipy.stats import nbinom

r = 6
p_succ = 0.35
x_vals = list(range(20))

cdf_vals = nbinom.cdf(x_vals, r, p_succ)

for val, prob in zip(x_vals, cdf_vals):
    print(f"CDF({val}) = {prob:.4f}")
```

Зобразіть графічно значення функції розподілу від'ємного біноміального розподілу з параметрами $n, p$ в точках вектора $x$.

```python
# Місце для Вашого коду
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.step(x_vals, cdf_vals, where='post')
plt.title(f"CDF нег. біноміального розподілу (r={r}, p={p_succ})")
plt.xlabel("x — кількість невдач до r успіхів")
plt.ylabel("F(X ≤ x)")
plt.grid(True)
plt.show()

```

### 1.3 Випадкові величини

Згенеруйте $100$ випадкових величин з від'ємного біноміального розподілу з параметрами $n, p$.

```python
# Місце для Вашого коду

from scipy.stats import nbinom

samples = [nbinom.rvs(r, p_succ) for _ in range(100)]

for value in samples:
    print(value, end=" ")
```

Виведіть кількість кожного елемента в отриманому випадковому векторі.

```python
# Місце для Вашого коду

freq = {}

for val in samples:
    freq[val] = freq.get(val, 0) + 1

for k, v in freq.items():
    print(f"count of {v} = {k}")
```

Зобразіть графічно кількість кожного елемента в отриманому випадковому векторі.

```python
# Місце для Вашого коду
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
x_vals = list(freq.keys())
y_vals = list(freq.values())
plt.stem(x_vals, y_vals, basefmt=" ")
plt.xlabel("Значення")
plt.ylabel("Кількість")
plt.title("Частота появи значень")
plt.grid(True)
plt.show()

```

### 3 Розподіл Пуассона

Зобразіть графічно функцію щільності, функцію розподілу та гістограму $10000$ випадкових величин розподілу Пуассона з параметрами:
1. $\mu = 0.1$
2. $\mu = 0.2$
3. $\mu = 0.3$
4. $\mu = 0.4$
5. $\mu = 0.5$
6. $\mu = 0.6$
7. $\mu = 0.7$
8. $\mu = 0.8$
9. $\mu = 0.9$

```python
# Місце для Вашого коду
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

params = np.arange(0.1, 1.0, 0.1)

for lam in params:
    values = np.arange(0, int(5 * lam) + 1)
    pmf = poisson.pmf(values, lam)
    cdf = poisson.cdf(values, lam)
    data = poisson.rvs(lam, size=10000)

    plt.figure(figsize=(10, 6))
    
    plt.hist(data, bins=values - 0.5, density=True, alpha=0.6, label="Histogram")
    plt.stem(values, pmf, basefmt=" ", linefmt='C1-', markerfmt='C1o', label="PMF")
    plt.step(values, cdf, where='post', color='green', label="CDF")

    plt.title(f"Poisson(λ = {lam:.1f}) — Histogram, PMF, CDF")
    plt.xlabel("k")
    plt.ylabel("Probability / Frequency")
    plt.legend()
    plt.grid(True)
    plt.show()

```

## 4 Апроксимація біноміального розподілу розподілом Пуассона

Покажіть схожість біноміального розподілу та розподілу Пуассона (їх функцій щільності) з відповідними параметрами.
Наприклад:
- $n = 100, p = 0.05$
- $n = 100, p = 0.1$
- $n = 100, p = 0.15$
- $n = 100, p = 0.2$
- $n = 100, p = 0.35$
- $n = 100, p = 0.5$
- $n = 100, p = 0.75$

```python
# Місце для Вашого коду
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom, poisson

trials = 100
prob_set = [0.05, 0.1, 0.15, 0.2, 0.35, 0.5, 0.75]

for prob in prob_set:
    rate = trials * prob
    k_vals = np.arange(0, int(rate + 4 * np.sqrt(rate)) + 1)

    bin_values = binom.pmf(k_vals, trials, prob)
    poi_values = poisson.pmf(k_vals, rate)

    plt.figure(figsize=(8, 4))
    plt.stem(k_vals, bin_values, basefmt=" ", label=f"Binomial(n={trials}, p={prob:.2f})")
    plt.stem(k_vals, poi_values, basefmt=" ", linefmt='C1-', markerfmt='C1o', label=f"Poisson(λ={rate:.2f})")
    plt.title(f"Порівняння: Binomial vs Poisson (p={prob:.2f}, λ={rate:.2f})")
    plt.xlabel("k")
    plt.ylabel("Ймовірність")
    plt.legend()
    plt.grid(True)
    plt.show()

```
